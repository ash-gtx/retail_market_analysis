```python
df = pd.read_excel('Online Retail.xlsx')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#finding percentage of null values
round(df.isnull().sum()/len(df)*100, 2)
```

```python
df.shape
```

```python
# removing rows with null values in Description
df = df.dropna(subset=['Description'])
df = df.dropna(subset=['CustomerID'])
df.shape
```

```python
round(df.isnull().sum()/len(df)*100, 2)
```

```python
np.median(df['CustomerID'])
```

```python
#quantity check min should be 0
df = df[df['Quantity']>=1]
df = df[df['UnitPrice']>=0.0]
```

```python
df.shape
```

```python
df
```

```python
df['Revenue'] = df['Quantity']*df['UnitPrice']
#keeping revenue upto 2 decimal places only
df['Revenue'] = df['Revenue'].round(2)

df.head()
```

```python
#extracting month from invoice date
df['InvoiceMonth'] = df['InvoiceDate'].dt.month
df.head()
```

```python
#exporting the file in csv format
df.to_csv('retail_cleaned.csv', index=False)
```
